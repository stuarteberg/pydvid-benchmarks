```python
import os
import glob
import collections
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
```

```python
from usecase_benchmarks import CSV_COLUMNS, get_empty_dataframe
```

```python
csv_files = glob.glob('/groups/flyem/proj/cluster/dvid_testing/*.csv')
print "Found {} files:".format( len(csv_files) )
for f in csv_files:
    print f
```

```python
tables = collections.OrderedDict()
for csv_file in csv_files:
    df = pd.io.parsers.read_csv(csv_file, index_col=False)
    df.insert(0, 'filename', os.path.split(csv_file)[1])
    tables[csv_file] = df
all_df = pd.concat(tables.values())
```

```python
all_df['cpu-seconds'] = all_df['elapsed-seconds'] * all_df['num-threads']
grouped = all_df.groupby('filename')
node_total = grouped['elapsed-seconds'].sum()

fig, ax = plt.subplots()
ind = np.arange(len(node_total))
ax.barh(ind, node_total)
ax.set_yticks(ind+0.5)
ax.set_yticklabels(node_total.keys())
ax.set_title('total node time');
```

```python
cpu_total =  grouped['cpu-seconds'].sum()
fig, ax = plt.subplots()
ind = np.arange(len(cpu_total))
ax.barh(ind, cpu_total)
ax.set_yticks(ind+0.5)
ax.set_yticklabels(cpu_total.keys())
ax.set_title('total cpu time');
```
